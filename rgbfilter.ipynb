```python
import numpy as np
from PIL import Image
import requests
import io
import matplotlib.pyplot as plt
```

```python
def imagevariants():
    url = input("Paste the URL of your image: ")
    image = np.asarray(Image.open(requests.get(url, stream=True).raw))

    #turns photo into array and takes needed colors

    def red():
        photo = np.asarray(Image.open(io.BytesIO(requests.get(url).content)))
        redphoto = photo.copy()
        redphoto[:, :, 1:] = 0
        return redphoto

    def green():
        photo = np.asarray(Image.open(io.BytesIO(requests.get(url).content)))
        greenphoto = photo.copy()
        greenphoto[:, :, 0] = 0
        greenphoto[:, :, 2] = 0
        return greenphoto

    def blue():
        photo = np.asarray(Image.open(io.BytesIO(requests.get(url).content)))
        bluephoto = photo.copy()
        bluephoto[:, :, 0:2] = 0
        return bluephoto

    def redgreen():
        photo = np.asarray(Image.open(io.BytesIO(requests.get(url).content)))
        redgreenphoto = photo.copy()
        redgreenphoto[:, :, 2] = 0
        return redgreenphoto

    def redblue():
        photo = np.asarray(Image.open(io.BytesIO(requests.get(url).content)))
        redbluephoto = photo.copy()
        redbluephoto[:, :, 1:2] = 0
        return redbluephoto

    def bluegreen():
        photo = np.asarray(Image.open(io.BytesIO(requests.get(url).content)))
        bluegreenphoto = photo.copy()
        bluegreenphoto[:, :, 0:1] = 0
        return bluegreenphoto

    redphoto = red()
    greenphoto = green()
    bluephoto = blue()
    redgreenphoto = redgreen()
    redbluephoto = redblue()
    bluegreenphoto = bluegreen()

    images = [image, redphoto, greenphoto, bluephoto, redgreenphoto, redbluephoto, bluegreenphoto]

    for ima in images:
        plt.figure()
        plt.imshow(ima)

imagevariants()
```
